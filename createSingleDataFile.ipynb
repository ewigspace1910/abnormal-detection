# COMPOSITE DATAFILE
 script to append all the csv data files into 1, require cleaned datafiles
 

1.   mount to my drive
2.   read all csv file in folder dataset
3.   append all into one 



```python
#mounted to drive 
#read and append

import sys
import pandas as pd
from sklearn.utils import shuffle

dataPath = "/content/drive/MyDrive/Colab Notebooks/abnormal-detection/cleanedDataset/"
filenames = open("/content/drive/MyDrive/Colab Notebooks/abnormal-detection/cleanedDataset/fnames.txt", "r")
names_set = [x.strip() for x in filenames if x[0] != '#']

try:
  df = pd.read_csv(dataPath + names_set[0])
  print(df.shape)
  for name in names_set[1:5]:
    print("Append ", name, " file")
    data = pd.read_csv(dataPath + name)
    df = df.append(data, ignore_index=True)
  print("new df shape :", df.shape)  

except Exception as e:
  print("error read file\n#########". e )
```

EDIT DATASET

```python
#random position row
df = shuffle(df) 
print(df.shape)
```

Make file output : csv file



```python

df.to_csv(dataPath + 'allinOneInput.csv', index=False)

#df.to_pickle(dataPath + 'allInoneInput.pickle')
```
