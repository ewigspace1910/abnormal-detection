```python
import csv
import sys
import datetime
from dateutil import parser #standardalize timedata
from collections import defaultdict

dirPath = "/content/drive/MyDrive/Colab Notebooks/dataset/" #importdata
save_dirPath = "/content/drive/MyDrive/Colab Notebooks/abnormal-detection/cleanedDataset/" #export data

                  
      
```

```python
#read and clean data function
def cleanData(inFile):
  '''
  cleanData(inFile, outFile)

  inFile : name of file in the datasetfolder
  outFile: name of file will be creat and save in cleanedDataset
  no return
  '''
  dropStats = defaultdict(int) 
  record = []
  print("cleaning {}".format(inFile))
  
        
  with open(inFile, 'r') as csvfile:
    data = csvfile.readlines()
    totalrows = len(data)
    print("total row read {}".format(totalrows))
    header = data[0]
            
    for line in data[1:] :
      line = line.strip()
      cols = line.split(",")
      key = cols[-1]
                
      #reduce line has a infinity or large value
      if line.startswith("D") or line.find("Infinity") >= 0 or line.find("infinity") >= 0:
        dropStats[key] += 1
        continue
                
      dt = parser.parse(cols[2]) # '1/3/18 8:17
      epochs = (dt - datetime.datetime(2000,1,1)).total_seconds()
      cols[2] = str(epochs)
      line = ','.join(cols)
  
      record.append(line)

    print("drop ->", dropStats )
    
  return header, record
```

```python
#write data to new file csv
def saveData(outFile, df, header):
  '''
  outFile : path to where file be saved from
  df : data frame
  header: index-column set
  '''
  with open(outFile, "w") as outfile:
        outfile.write(header.strip() + "\n")
        for line in df:
          outfile.write(line + "\n")
```

Execute over all file

```python
list_files = open(dirPath + "fnames.txt")
names_file = []
for f in list_files:
  f = f.strip()
  if f[0] != '#' :
    header, df = cleanData(dirPath + f)
    saveData(save_dirPath + f, df, header)
    names_file.append(f)

list_files_save = open(save_dirPath + "fnames.txt", "w")
list_files_save.write(names_file)
```
